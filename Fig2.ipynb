```python
#b
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

plt.rcParams['text.usetex'] = False
expr_df = adata_aucs.to_df()
expr_df['major_celltype'] = adata_aucs.obs['major_celltype'].values
mean_expr = expr_df.groupby('major_celltype').mean().T
target_genes = [
 "positive regulation of smoothened signaling pathway(GO:0045880)",
"protein-glutamine gamma-glutamyltransferase activity(GO:0003810)",
"mRNA splicing, via spliceosome(GO:0000398)",
"negative regulation of cell growth(GO:0030308)",
"ribosomal large subunit assembly(GO:0000027)",
"G1/S transition of mitotic cell cycle(GO:0000082)",
"ribosome binding(GO:0043022)",
"positive regulation of cell population proliferation(GO:0008284)",
"ribosomal large subunit biogenesis(GO:0042273)",
"defense response to other organism(GO:0098542)",
"catechol oxidase activity(GO:0004097)",
"melanotic encapsulation of foreign target(GO:0035011)",
"antioxidant activity(GO:0016209)",
"hemolymph coagulation(GO:0042381)",
"response to wounding(GO:0009611)",
"positive regulation of antimicrobial peptide biosynthetic process(GO:0002807)",
"induction of bacterial agglutination(GO:0043152)",
"antifungal innate immune response(GO:0061760)",
"positive regulation of lysosomal protein catabolic process(GO:1905167)",
"CD40 signaling pathway(GO:0023035)",
"positive regulation of receptor internalization(GO:0002092)",
"pigment granule maturation(GO:0048757)",
"positive regulation of engulfment of apoptotic cell(GO:1901076)",
"reactive nitrogen species metabolic process(GO:2001057)",
"late endosome(GO:0005770)",
"cellular response to reactive oxygen species(GO:0034614)",
]
mean_expr_sub = mean_expr.loc[target_genes]
mean_expr_norm = mean_expr_sub.apply(lambda x: (x - x.mean()) / x.std(), axis=1)
df=mean_expr_norm

col_colors = {
    'CTL Hemocytes': '#81bfdd',
    'Crustin Hemocytes': '#3e8dab',
    'TGase Hemocytes': '#f19795'
}
col_color_list = [col_colors[col] for col in df.columns]

figsize = (1.8, 12)
g = sns.clustermap(df,
                   cmap='Reds',
                   annot=False,
                   col_colors=col_color_list,
                   row_cluster=False,
                   col_cluster=False,
                   figsize=figsize)
g.ax_heatmap.grid(False)
g.ax_heatmap.set_xlabel('')
g.ax_heatmap.set_xticklabels(g.ax_heatmap.get_xticklabels(),
                             rotation=30,
                             ha='right',
                             fontsize=8)
g.ax_heatmap.set_ylabel('')
g.ax_heatmap.set_yticklabels(g.ax_heatmap.get_yticklabels(), fontsize=12)
g.cax.set_position([-0.2, 0.4, 0.2, 0.3])
g.ax_col_colors.set_position([0.39, 0.67, 0.57, 0.01]) 
g.ax_col_dendrogram.set_position([0.01, 0, 0.01, 0.01])

plt.show()
```
