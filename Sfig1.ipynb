```python
#b
samples <- read.table("./samples2.txt", sep='\t',header = FALSE, stringsAsFactors = FALSE)
df <- dcast(samples, V2 ~ V1, value.var = "V1", fill = 0)
rownames(df) <- df$V2
df <- df[, -1] 
df[is.na(df)] <- 0
df[df != 0] <- 1
traitData <- df   
WGCNA_matrix <- datExpr0  

commonSamples <- intersect(rownames(WGCNA_matrix), rownames(traitData))
datExpr_sub   <- WGCNA_matrix[commonSamples, ]
datTraits     <- traitData[commonSamples, ]
stopifnot(identical(rownames(datExpr_sub), rownames(datTraits)))

nSamples <- nrow(datExpr_sub)
moduleTraitCor    <- cor(mergedMEs[commonSamples, ], datTraits, use = "p")
moduleTraitPvalue <- corPvalueStudent(moduleTraitCor, nSamples)

sizeGrWindow(10, 6)
textMatrix <- paste(signif(moduleTraitCor, 2), "\n(",
                    signif(moduleTraitPvalue, 1), ")", sep = "")
dim(textMatrix) <- dim(moduleTraitCor)
colors <- colorRampPalette(c("#FFFFCC", "white", "#800080"))(20)
png("heatmap.png", width = 5, height = 7, units = "in", res = 600, family = "Arial")
par(mar = c(6, 10, 1, 2))
labeledHeatmap(Matrix = moduleTraitCor,
               xLabels = names(datTraits),
               yLabels = names(mergedMEs),
               ySymbols = names(mergedMEs),
               colorLabels = FALSE,
               colors = colors,
               # textMatrix = textMatrix,  
               setStdMargins = FALSE,
               cex.text = 0.4,
               cex.lab = 0.8,
               zlim = c(-1,1),
               main = paste(""))
dev.off()
```
