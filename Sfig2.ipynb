```python
#2a
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['font.family'] = 'Arial'

adata=ov.read('umap3.h5ad')
sc.pp.normalize_total(adata)
sc.pp.log1p(adata)

import pandas as pd
tfgenes = pd.read_csv('unchgene.txt', header=None, sep='\t')
tfgenes_list = tfgenes[0].values  
valid_genes = pd.Index(tfgenes_list).intersection(adata.var.index)
tfdata = adata[:, valid_genes].copy()

sc.tl.rank_genes_groups(
    tfdata, groupby="major_celltype", use_raw=False,
    method="t-test", key_added="dea_leiden_res1"
)

sc.pl.rank_genes_groups_dotplot(
    tfdata, groupby="major_celltype", swap_axes=False,
    standard_scale="var", n_genes=3, key="dea_leiden_res1",dendrogram=False,show=False
)
```

```python
#2b
import re
sel_genes = adata.var_names[adata.var_names.str.match(r'^(LNCG|CODG)')]
adata_sub = adata[:, sel_genes].copy()
sc.tl.rank_genes_groups(
    adata_sub, groupby="major_celltype", use_raw=False,
    method="t-test", key_added="dea_leiden_res1"
)
sc.tl.filter_rank_genes_groups(
    adata_sub,
    min_in_group_fraction=0.2,
    max_out_group_fraction=0.2,
    key="dea_leiden_res1",
    key_added="dea_leiden_res1_filtered",
)

sc.pl.rank_genes_groups_dotplot(
    adata_sub, groupby="major_celltype", swap_axes=False,cmap='Blues',
    standard_scale="var", n_genes=3, key="dea_leiden_res1_filtered",dendrogram=False,show=False
)
```
