```python
#h
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from sklearn.preprocessing import StandardScaler
tpm = pd.read_csv("/public5/xiongxi2/Shrimpbase/Vannamei/PLOT/modified_tmm", sep='\t',index_col=0)
tpm = tpm.loc[["gene-LOC113817261",
"gene-LOC113817723",
"gene-LOC113802097",
"gene-LOC113826098",
"gene-LOC113817260",
"gene-LOC113805736",
"gene-LOC113805737",
"gene-LOC113805758",
"gene-LOC113817089",
"gene-LOC113803089",
"gene-LOC113814399",
"gene-LOC113812940",
"gene-LOC113822166",
"gene-LOC113825851"], :]

def assign_group(sample_name):
    if '_stage' in sample_name.lower():
        return 'Larva'

    elif '_molt' in sample_name.lower():
        return 'Molting-Whole body'
    elif sample_name.startswith('Hep'):
        return 'Hepatopancreas'
    elif sample_name.startswith('I'):
        return 'Intestine'
    elif sample_name.startswith('Mix'):
        return 'Hepatopancreas'
    elif sample_name.startswith('Carapace'):
        return 'Hepatopancreas'
    else:
        return 'Other tissue'


sample_groups = pd.Series({col: assign_group(col) for col in tpm.columns}, name='Tissue')
group_info = sample_groups.to_frame()
group_info.index.name = 'Sample'
lut = {
    'Hepatopancreas': '#E64B35',
    'Intestine': '#4DBBD5',
    'Larva': '#a8d5a2',
    'Molting-Whole body': '#A569BD',
    'Other tissue': '#F7DC6F'
}
col_colors = group_info['Tissue'].map(lut)
group_order = ['Hepatopancreas', 'Intestine',  
               'Larva',
               'Molting-Whole body', 'Other tissue']
sorted_samples = group_info.reset_index().sort_values(
    by='Tissue', 
    key=lambda x: x.map({k: i for i, k in enumerate(group_order)})
)['Sample'].tolist()

group_order_dict = {grp: i for i, grp in enumerate(group_order)}

group_info = sample_groups.to_frame(name='Tissue')
group_info.index.name = 'Sample'
group_info['group_order'] = group_info['Tissue'].map(group_order_dict)
group_info_sorted = group_info.sort_values(by='group_order')
sorted_samples = group_info_sorted.index.tolist()
tpm_sorted = tpm[sorted_samples]
col_colors_sorted = col_colors.loc[sorted_samples]
tpm_sorted.index = tpm_sorted.index.str.replace(r"^gene-", "", regex=True)

rename_map = {
"LOC113817261":"CHIA",
"LOC113817260":"CHIT1",
"LOC113803089":"Endochitinase",
"LOC113817723":"AMY1A",
"LOC113802097":"CPB1.1",
"LOC113826098":"CPB1.2",
"LOC113805758":"CTRB1.1",
"LOC113817089":"CTRB1.2",
"LOC113805736":"CTRB2.1",
"LOC113805737":"CTRB2.2",
"LOC113814399":"EST4.1",
"LOC113812940":"EST4.2",
"LOC113822166":"SCD5",
"LOC113825851":"PRSS1.1"}


tpm_sorted.index = tpm_sorted.index.map(lambda x: rename_map.get(x, x))


g = sns.clustermap(tpm_sorted,
                   col_colors=col_colors_sorted,
                   col_cluster=False,
                   row_cluster=False,
                   cmap='YlGnBu',
                   figsize=(9, 4.5),
                   cbar_kws={'label': 'TPM'},
                   xticklabels=False,
                   yticklabels=True,
                   vmin=0,
                   vmax=1000)

handles = [mpatches.Patch(color=color, label=label) for label, color in lut.items()]
g.ax_col_dendrogram.legend(handles=handles,
                           title='Tissue',
                           bbox_to_anchor=(-0.3, 0.01),
                           loc='upper left',
                           borderaxespad=0.)

g.ax_heatmap.set_xticklabels(g.ax_heatmap.get_xticklabels(), fontsize=3)
g.cax.set_position([0.04, 0.1, 0.04, 0.3]) 
plt.show()
```
