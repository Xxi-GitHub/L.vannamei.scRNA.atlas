```python
# Here is a sample
group_info = pd.DataFrame({
    'Sample': ['SRR6347878', 'SRR6347879',	
                  'SRR6347880', 'SRR6347881',
                  'SRR6347882', 'SRR6347883'
],
    'Group':  ['normal', 'normal', 'normal', 'infection', 'infection', 'infection',]
}).set_index('Sample')
tpm_selected = tpm_selected[group_info.index]
tpm_selected = tpm_selected[(tpm_selected >= 3).sum(axis=1) >= 2]
```

```python
lut = {'normal': '#4DBBD5', 'infection': '#E64B35'}
col_colors = group_info['Group'].map(lut)
tpm_scaled = pd.DataFrame(
    StandardScaler().fit_transform(tpm_selected.T).T,
    index=tpm_selected.index,
    columns=tpm_selected.columns
)

cell_width = 1
cell_height = 1
n_genes, n_samples = tpm_scaled.shape
figsize = (5,5)
g = sns.clustermap(
    tpm_scaled,
    col_colors=col_colors,
    col_cluster=False,
    row_cluster=False,
    cmap='vlag',
    figsize=figsize,
    cbar_kws={'label': 'TPM Z-score'},
    xticklabels=True,
    yticklabels=True
)
g.fig.suptitle("PRJNA421143-Hepatopancreas-Vibrio parahaemolyticus", fontsize=12, y=1.02)
handles = [mpatches.Patch(color=color, label=label) for label, color in lut.items()]
g.ax_col_dendrogram.legend(handles=handles,
                           title='Group',
                           loc='center',
                           bbox_to_anchor=(0.5, 0.4),
                           ncol=2,
                           fontsize=12,
                           title_fontsize=12,
                           frameon=False)

for i, gene in enumerate(tpm_scaled.index):
    for j, sample in enumerate(tpm_scaled.columns):
        val = tpm_selected.loc[gene, sample]
        g.ax_heatmap.text(j + 0.5, i + 0.5,
                          f'{val:.1f}',
                          ha='center', va='center',
                          fontsize=10, color='black')
plt.rcParams['font.family'] = 'Arial'

plt.show()
```
